```python
import pandas as pd
import ipywidgets as widgets
from itables import show
from IPython.display import display, clear_output
```

```python
# Load dataset
df = pd.read_csv('tl_items.csv')

# Drop specified columns
drop_cols = ['value', 'Survey_Name', 'Item_Number', 'Survey_Item_Number']
df = df.drop(columns=drop_cols, errors='ignore')

# Move the last column to the first position
cols = df.columns.tolist()
df = df[[cols[-1]] + cols[:-1]]

# 1. Filter columns: Only keep columns where values repeat
# A column is excluded if every row has a unique value
filter_columns = [
    col for col in df.columns 
    if df[col].nunique() < len(df) and col.lower() != 'id'
]

# 2. Create UI Widgets
filter_widgets = {}
for col in filter_columns:
    options = ['All'] + sorted(df[col].dropna().unique().astype(str).tolist())
    filter_widgets[col] = widgets.Dropdown(
        options=options,
        value='All',
        description=f'{col}:',
        layout=widgets.Layout(width='auto', min_width='200px')
    )

# 3. Define the update function
def update_view(*args):
    clear_output(wait=True)
    display(ui_layout)
    
    filtered_df = df.copy()
    for col, widget in filter_widgets.items():
        if widget.value != 'All':
            filtered_df = filtered_df[filtered_df[col].astype(str) == widget.value]
    
    # Calculate indices for all columns except the first one (Item)
    meta_indices = list(range(1, len(filtered_df.columns)))

    show(
        filtered_df, 
        classes="display", 
        autoWidth=False,
        columnDefs=[
            {"className": "dt-left", "targets": "_all"},
            {"width": "1%", "targets": meta_indices}
        ]
    )

# 4. Link widgets
for widget in filter_widgets.values():
    widget.observe(update_view, 'value')

# 5. Condense Layout using GridBox
# This creates a 3-column responsive grid
ui_layout = widgets.GridBox(
    list(filter_widgets.values()),
    layout=widgets.Layout(
        grid_template_columns='repeat(3, 1fr)',
        grid_gap='10px',
        padding='10px',
        border='1px solid #ddd'
    )
)
```
